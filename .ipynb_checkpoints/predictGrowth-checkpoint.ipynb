```python
full_df = pd.read_csv("./data/clean/cleanDataLong.csv")
cleanDf = full_df.dropna(axis=0)
print cleanDf.shape
cleanDf.head()
```

```python
sns.regplot(x="monthToNow", y="price", data=cleanDf, x_estimator=np.mean)
sns.plt.xlim(0,250)
sns.plt.ylim(0,1500000)
plt.title("value trend - all data")
```

```python
sns.lmplot(x="monthToNow", y="price", data=cleanDf, x_estimator=np.mean, scatter=False, hue="city")
sns.plt.xlim(0,250)
sns.plt.ylim(0,)
plt.title("value trend - by city")
```

```python
sns.lmplot(x="monthToNow", y="price", data=cleanDf, x_estimator=np.mean, scatter=False, hue="type")
sns.plt.xlim(0,250)
sns.plt.ylim(0,)
plt.title("value trend - by type")
```
